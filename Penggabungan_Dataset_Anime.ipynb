```python
import pandas as pd

df1 = pd.read_csv('/content/animes.csv')
df2 = pd.read_csv('/content/top_anime_dataset.csv')
```

```python
print(df1.shape, df2.shape)
```

```python
df1 = df1[df1['genre'].apply(lambda x: 'Hentai' not in x)]
```

```python
df1.rename(columns={"title": "anime_title"}, inplace=True)
df2.rename(columns={"Name": "anime_title"}, inplace=True)
```

```python
print(df1.columns)
print(df2.columns)
```

```python
final_df = pd.merge(df1, df2, on="anime_title", how="inner")
```

```python
final_df.columns
```

```python
print(final_df.isnull().sum())
```

```python
print(final_df.shape)
```

```python
final_df.drop_duplicates(inplace=True)
```

```python
print(final_df.shape)
```

```python
final_df = final_df.drop(columns=["popularity", "Members", "Ranked"])
```

```python
final_df["score"] = final_df["score"].astype(float)
```

```python
final_df['score'] = final_df['score'].fillna(final_df['Rating'])
```

```python
print(final_df.isnull().sum())
```

```python
final_df.to_csv('merged_df.csv', index=False)
```
